```python
import faiss
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore
```

```python
#len = len(query_embedding)
#print(len)
index = faiss.IndexFlatL2(384)
```

```python
vector_store=FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={}
)
```

```python
vector_store.add_texts(["AI is future", "AI is powerful","Dogs are cute"])
```

```python
vector_store.index_to_docstore_id
```

```python
vector_store.similarity_search("Tell me about AI", k=3)
```

### Veeras example

```python
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
import numpy as np

A = np.array([[1, 1, 1, 0, 0]])
B = np.array([[0, 0, 1, 1, 1]])

cos_sim = cosine_similarity(A, B)
euc_dist = euclidean_distances(A, B)

print("Cosine Similarity:", cos_sim[0][0])    # e.g., 0.333...
print("Euclidean Distance:", euc_dist[0][0])  # e.g., 2.0
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances

docs = ["AI is transforming healthcare", "AI revolutionizes the medical industry"]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)

cos_sim = cosine_similarity(X)[0, 1]
euc_dist = euclidean_distances(X)[0, 1]

print("Cosine Similarity:", round(cos_sim, 3))
print("Euclidean Distance:", round(euc_dist, 3))
```
